# Songs for artist

```python
def get_primary_artist(song_json):
    return song_json["primary_artist"]["name"]

def fetch_page(artist_id, page):
    songs = []
    base_url = "https://api.genius.com/artists/{}/songs".format(artist_id)
    params = {
        "per_page": 30,
        "page": page,
        "sort": "popularity",
        "access_token": "N/A"
    }
    response = requests.get(base_url, params=params).json()["response"]
    for song in response["songs"]:
        primary_artist = song["primary_artist"]["name"]
        if primary_artist == "NAV":
            songs.append(unicodedata.normalize("NFKD", song["full_title"]))
    
    return songs, response["next_page"]

def collect_songs(artist_id):
    all_songs = []
    page = 1
    while True:
        ret, next_page = fetch_page(artist_id, page)
        all_songs.extend(ret)
        
        if next_page is None:
            break
        page = next_page
    
    return all_songs
    
```
