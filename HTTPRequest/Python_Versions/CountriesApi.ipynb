```python
import requests
import json
```

```python
base_url = "https://restcountries.eu/rest/v2/name/united"
params = {}

resp = requests.get(base_url, params=params).json()
```

```python
resp
```

```python
def build_country_obj(country_json):
    country_obj = {}
    country_obj["name"] = country_json["name"]
    
    # Get currencies
    all_currencies = []
    for currency in country_json["currencies"]:
        all_currencies.append(currency["name"])
    country_obj["currencies"] = all_currencies
    
    # Get languages
    all_languages = []
    for language in country_json["languages"]:
        all_languages.append(language["name"])
    country_obj["languages"] = all_languages
    
    # Get populations
    country_obj["population"] = country_json["population"]
    
    return country_obj
```

```python
all_new_objects = []
for elem in resp:
    all_new_objects.append(build_country_obj(elem))
#     print("_________________________________________________")
```
