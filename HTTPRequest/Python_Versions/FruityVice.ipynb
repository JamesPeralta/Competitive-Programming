```python
base_url = "https://www.fruityvice.com/api/fruit/all"
params = {}
resp = requests.get(base_url, params=params).json()
print(resp)
```

```python
print(json.dumps(resp, indent=2, sort_keys=True))
```

```python
result = []
for fruit in resp:
    result.append(create_fruit_obj(fruit))
```

```python
def create_fruit_obj(fruit_json):
    fruit_obj = {}
    fruit_obj["name"] = fruit_json["name"]
    fruit_obj["calories"] = fruit_json["nutritions"]["calories"]
    fruit_obj["carbohydrates"] = fruit_json["nutritions"]["carbohydrates"]
    fruit_obj["fat"] = fruit_json["nutritions"]["fat"]
    fruit_obj["protein"] = fruit_json["nutritions"]["protein"]
    fruit_obj["sugar"] = fruit_json["nutritions"]["sugar"]
    
    return fruit_obj
```
